```julia
using StatsBase, DataFrames, DataFrameMacros
using RCall, CategoricalArrays, TexTables
using Distributions, Gadfly, Compose, MLJ
using StableRNGs
import DecisionTree

using MLJ: schema, Continuous
```

```julia
include("pubh.jl")
set_default_plot_size(15cm, 12cm)
```

# Data

```julia
iris = rcopy(R"datasets::iris")
iris |> schema
```

We `unpack` the data to define *response* and *explanatory* variables:

```julia
y, X = unpack(iris, ==(:Species));
X |> head
```

## Data sppliting

```julia
rng = StableRNG(566)
train, test = partition(eachindex(y), 0.7, shuffle=true, rng=rng);
```

# Pruned Trees

## Model

```julia
using MLJDecisionTreeInterface
```

```julia
dtc = @load DecisionTreeClassifier pkg=DecisionTree verbosity=0
pruned_model = dtc(max_depth=2)
pruned_mach = machine(pruned_model, X, y)
```

```julia
fit!(pruned_mach, rows=train);
```

## Visualisation

```julia
fitted_params(pruned_mach).tree
```

```julia
pruned_mach |> feature_importances |> DataFrame
```

## Performance

```julia
pruned_yÃÇ = MLJ.predict(pruned_mach, rows=test);
pruned_yÃÑ = predict_mode(pruned_mach, rows=test);
```

```julia
cross_entropy(pruned_yÃÇ, y[test]) |> mean |> r3
```

```julia
misclassification_rate(pruned_yÃÑ, y[test]) |> r3
```

```julia
accuracy(pruned_yÃÑ, y[test]) |> r3
```

```julia
kappa(pruned_yÃÑ, y[test]) |> r3
```

When we have many categories, it is easy to construct the data frame, and from that data frame, create binary variables and then construct the corresponding confussion matrix.

```julia
prune_df = DataFrame(
  hcat(
    string.(pruned_yÃÑ),
    string.(y[test])
  ),
  [:pred, :obs]
);
```

```julia
prune_df.pred_bin = prune_df.pred .== "versicolor"
prune_df.obs_bin = prune_df.obs .== "versicolor";
```

```julia
@rlibrary epitools
epitable(prune_df.pred_bin, prune_df.obs_bin, rev="both")
```

```julia
sensitivity(prune_df.pred_bin, prune_df.obs_bin) |> r3
```

```julia
specificity(prune_df.pred_bin, prune_df.obs_bin) |> r3
```

# Random Forests

## Model

```julia
rfc = @load RandomForestClassifier pkg=DecisionTree verbosity=0
forest = rfc(
  	n_subfeatures = 2,
  	n_trees = 10,
  	sampling_fraction = 0.5,
  	max_depth = 6
)
forest_mach = machine(forest, X, y)
```

```julia
fit!(forest_mach, rows=train);
```

```julia
fitted_params(forest_mach)
```

```julia
forest_mach |> feature_importances |> DataFrame
```

## Performance

# Linear Discriminant Analysis

## Model

```julia
using MLJMultivariateStatsInterface
```

```julia
LDA = @load LDA pkg=MultivariateStats verbosity=0
lda_model = LDA()
lda_mach = machine(lda_model, X, y)
```

```julia
fit!(lda_mach, rows=train);
```

## Visualisation

```julia
lda_x = MLJ.transform(lda_mach)
lda_x |> size
```

```julia
lda_x |> head
```

```julia
lda_df = DataFrame(
	PC1 = lda_x.x1,
	PC2 = lda_x.x2,
	Species = string.(y)
)

lda_df |> head
```

```julia
#| label: fig-fig1
#| fig-cap: "Principal components used to classify the iris data set."
plot(
  lda_df,
  x=:PC1, y=:PC2, color=:Species,
  Geom.point,
  Guide.xlabel("PC1"),
  Guide.ylabel("PC2")
)
```

## Performance

```julia
lda_yÃÇ = MLJ.predict(lda_mach, rows=test);
lda_yÃÑ = predict_mode(lda_mach, rows=test);
```

```julia
cross_entropy(lda_yÃÇ, y[test]) |> mean |> r3
```

```julia
misclassification_rate(lda_yÃÑ, y[test]) |> r3
```

```julia
accuracy(lda_yÃÑ, y[test]) |> r3
```

```julia
kappa(lda_yÃÑ, y[test]) |> r3
```
